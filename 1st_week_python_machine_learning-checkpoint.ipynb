## Feature Selection

```python
corr = data.corr()
plt.figure(figsize=(18,18))
sns.heatmap(data.corr(), cmap='coolwarm', annot = True)
plt.title("Correlation Map", fontweight = "bold", fontsize=16)
```

```python
#dropping the feature that as high correlation with more than one feature

drop_corr = data.drop(['perimeter_mean','radius_mean','compactness_mean','concave points_mean','radius_se','perimeter_se',
                 'radius_worst','perimeter_worst','compactness_worst','concave points_worst','compactness_se',
                'concave points_se','texture_worst','area_worst'], axis =1)
drop_corr.head()
                
```

```python
#Checking the correlation of selected ones

corr = drop_corr.corr()
plt.figure(figsize=(18,18))
sns.heatmap(drop_corr.corr(), cmap='coolwarm', annot = True)
plt.title("Correlation Map", fontweight = "bold", fontsize=16)
```

## Prerequisties for using this algorithms 

- Important features that contirbutes massively to a more accurate prediction
- Filtering out the unnecessary features.

```python
prediction_var = list(drop_corr.drop('diagnosis', axis=1))
```

```python
prediction_var
```

```python
train, test = train_test_split(data, test_size = 0.3)
```

```python
train_X = train[prediction_var]
train_Y = train.diagnosis

test_X = test[prediction_var]
test_Y = test.diagnosis
```

## Random forest Classifer.

In simple word, it is ensemble of decision tree.

- Each tree gives a decision
- Tree with most 'votes' = correct prediciton ( eg. There are 7 trees, 4 say bengeine, then its begine)

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
prediction
```

```python
metrics.accuracy_score(prediction, test_Y)
```

## Support Vector Machine

- Optimal hyperpane in an N-dimensional space(N â€” the number of features) that distinctly classifies the data points.

```python
model = svm.SVC()
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
metrics.accuracy_score(prediction, test_Y)
```

## Linear Regression

- determine a line which best fits the data

```python
model = LogisticRegression()
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
metrics.accuracy_score(prediction, test_Y)
```
